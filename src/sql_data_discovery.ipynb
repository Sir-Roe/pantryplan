```python
n = "recipe_ingredients.csv"
df_ing=pd.read_csv(f'{sql_dir}\{n}')
df_ing['quantity']=df_ing['quantity'].astype(str)
```

```python
value_to_drop = 'toppings of choice, e.g. maple syrup or fruit, for serving'
df_ing = df_ing[~(df_ing['quantity'] == value_to_drop)]
df_ing=df_ing.reset_index(drop=True)
df_ing=df_ing.replace('nan','')
```

```python
df_ing['quantity']
```

```python
for index, row in df_ing.iterrows():
    quantity = row['quantity'].strip()  # Remove leading/trailing spaces
    if " to " in quantity:  # Check if " to " is present
        range_parts = quantity.split(" to ")
        a = range_parts[0].strip()
        b = range_parts[1].strip()
        
        # Check if a contains fractions
        if "/" in a:
            a = a.split()
            s = 0
            for val in a:
                if "/" in val:
                    frac_a = Fraction(val)
                    s += frac_a
                else:
                    s += int(val.strip())
        else:
            s = float(a)
        a = float(s)
        
        # Check if b contains fractions
        if "/" in b:
            b = b.split()
            s = 0
            for val in b:
                if "/" in val:
                    frac_b = Fraction(val)
                    s += frac_b
                else:
                    s += int(val.strip())
        else:
            s = float(b)
        b = float(s)
        
        q = (b + a) / 2
        df_ing.at[index, "quantity"] = q
        print([row['quantity'], "original ^", q])
```
